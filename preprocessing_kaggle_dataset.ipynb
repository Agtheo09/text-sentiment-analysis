```python
import pandas as pd

train_df = pd.read_csv('./data/kaggle_dataset/train_kaggle.csv')
test_df = pd.read_csv('./data/kaggle_dataset/test_kaggle.csv')

train_df = train_df[['text', 'sentiment']]
test_df = test_df[['text', 'sentiment']]

train_df['text'] = train_df['text'].fillna("")
test_df['text'] = test_df['text'].fillna("")

train_df['text'] = train_df['text'].apply(lambda x: x.encode('utf-8', 'ignore').decode('utf-8'))
test_df['text'] = test_df['text'].apply(lambda x: x.encode('utf-8', 'ignore').decode('utf-8'))

# remove links from the text column
train_df['text'] = train_df['text'].str.replace(r'http\S+', '', regex=True)
test_df['text'] = test_df['text'].str.replace(r'http\S+', '', regex=True)

train_df['text'] = train_df['text'].fillna("")
test_df['text'] = test_df['text'].fillna("")

train_positive_df = train_df[train_df['sentiment'] == 'positive'].sample(7700, random_state=42)
train_neutral_df = train_df[train_df['sentiment'] == 'neutral'].sample(7700, random_state=42)
train_negative_df = train_df[train_df['sentiment'] == 'negative'].sample(7700, random_state=42)

test_positive_df = test_df[test_df['sentiment'] == 'positive'].sample(1000, random_state=42)
test_neutral_df = test_df[test_df['sentiment'] == 'neutral'].sample(1000, random_state=42)
test_negative_df = test_df[test_df['sentiment'] == 'negative'].sample(1000, random_state=42)

train_balanced_df = pd.concat([train_positive_df, train_neutral_df, train_negative_df])
test_balanced_df = pd.concat([test_positive_df, test_neutral_df, test_negative_df])

# instead of positive neutral negative, we will use 1 0 -1 for the sentiment values use 0 negative 1 for positive and 2 for neutral
train_balanced_df['sentiment'] = train_balanced_df['sentiment'].map({'positive': 1, 'neutral': 2, 'negative': 0})
test_balanced_df['sentiment'] = test_balanced_df['sentiment'].map({'positive': 1, 'neutral': 2, 'negative': 0})

train_balanced_df.to_csv('./data/kaggle_dataset/train_balanced_kaggle.csv', index=False)
test_balanced_df.to_csv('./data/kaggle_dataset/test_balanced_kaggle.csv', index=False)
```
